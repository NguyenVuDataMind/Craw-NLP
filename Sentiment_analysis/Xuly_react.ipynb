```python
import re
from pymongo import MongoClient

# Kết nối tới MongoDB
client = MongoClient('mongodb+srv://MLadmin:admin1021@machinelearning.so3qxxp.mongodb.net/')
db = client['KPW']
col_group = db['groupScraping']

# Các loại react tích cực và tiêu cực
positive_reacts = ['Thích', 'Wow', 'Yêu thích', 'Haha', 'Thương thương']
negative_reacts = ['Buồn', 'Phẫn nộ']

# Hàm trích xuất số lượng react từ chuỗi "Hiển thị X người đã bày tỏ cảm xúc"
def extract_react_count(react_string):
    # Kiểm tra nếu react là "Tất cả", nếu có thì trả về 0 (bỏ qua)
    if "Tất cả" in react_string:
        return 0
    
    # Sử dụng biểu thức chính quy để trích xuất số lượng react
    match = re.search(r'Hiển thị (\d+)', react_string)
    if match:
        return int(match.group(1))
    return 0

# Hàm tính Sentiment Score
def calculate_sentiment_score(reacts):
    # Khởi tạo số lượng react tích cực và tiêu cực
    positive_count = 0
    negative_count = 0
    total_count = 0
    
    # Duyệt qua các react trong danh sách
    for react in reacts:
        # Trích xuất số lượng react từ chuỗi
        count = extract_react_count(react)
        
        # Nếu react là "Tất cả", bỏ qua
        if count == 0:
            continue
        
        # Cộng vào số lượng react tích cực hoặc tiêu cực
        total_count += count
        if any(pos_react in react for pos_react in positive_reacts):
            positive_count += count
        elif any(neg_react in react for neg_react in negative_reacts):
            negative_count += count
    
    # Kiểm tra trường hợp không có react nào
    if total_count == 0:
        return 0  # Nếu không có react nào, trả về Sentiment Score là 0
    
    # Tính Sentiment Score theo công thức
    sentiment_score = (positive_count - negative_count) / total_count
    return sentiment_score

# Lấy dữ liệu từ MongoDB và tính Sentiment Score
for item in col_group.find():
    # In nội dung bài viết
    print(item['Text-Content'])
    
    # Lấy ra danh sách react
    reacts = item.get('Likedetails', [])
    
    # Tính toán Sentiment Score cho react của bài viết
    sentiment_score = calculate_sentiment_score(reacts)
    
    # In các react trong bài viết
    for i in item['Likedetails']:
        print(i)

    # In thêm Sentiment Score mà không làm ảnh hưởng tới các giá trị in ra khác
    print(f"Sentiment Score: {sentiment_score}")
    
    print('----------------------------------------------------------------------------------------')
```
