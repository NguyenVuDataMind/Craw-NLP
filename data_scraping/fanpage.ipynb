### Đăng nhập dùng cookie

```python
from selenium import webdriver
import json
import time
from selenium import webdriver
from time import sleep
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
# Tải cookie từ file
with open("facebook_cookies.json", "r") as file:
    cookies = json.load(file)

# Mở trình duyệt và vào Facebook
driver = webdriver.Chrome()
driver.get("https://www.facebook.com/dueconfessionsofficial")
# driver.get("https://www.facebook.com/aow.page")

time.sleep(2)  # Đợi trang tải

# Thêm cookie vào trình duyệt
for cookie in cookies:
    driver.add_cookie(cookie)

# Reload trang để áp dụng cookie
driver.refresh()

# Kiểm tra xem đã đăng nhập chưa
print("Đăng nhập thành công nếu thấy tài khoản của bạn!")
sleep(5)
```
